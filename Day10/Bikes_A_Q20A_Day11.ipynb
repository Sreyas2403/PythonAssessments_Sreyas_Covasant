```python
station_cols = ['Berri1', 'CSC', 'Mais1', 'Mais2', 'Parc', 'PierDup', 'Rachel1', 'Totem_Laurier']

f_df['total_attendance'] = f_df[station_cols].sum(axis=1)
weekday_totals = f_df.groupby('Weekday_name', as_index=False)['total_attendance'].sum()
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_totals['Weekday_name'] = pd.Categorical(
    weekday_totals['Weekday_name'], 
    categories=weekday_order, 
    ordered=True
)
weekday_totals = weekday_totals.sort_values('Weekday_name').reset_index(drop=True)
print(weekday_totals)
```
