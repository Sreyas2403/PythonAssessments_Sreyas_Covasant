```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft,fftfreq 
```

```python
duration =1 
sampling_rate = 1000
t = np.linspace(0,duration, int(duration*sampling_rate))
t
```

```python
frq1 = 5
frq2 = 50
```

```python
signal1 = np.sin(2 * (np.pi)* t * frq1)
signal2 = np.sin(2* (np.pi) * t * frq2)
```

```python
total_signal = signal1 + signal2
```

```python
total_signal
```

```python
n = int(len(t))
n
```

```python
xf = fftfreq(n, 1/sampling_rate)
yf = fft(total_signal)
```

```python
xf
```

```python
yf
```

```python
xfeq = xf[:n // 2]     #taking +ve values for fft components
yfeq = np.abs(yf[:n // 2])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(xfeq, yfeq)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.title("FFT of 5Hz and 50 Hz signals")
plt.show()
```
