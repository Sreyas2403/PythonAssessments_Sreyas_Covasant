```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n1= np.random.normal(loc=0,scale =1, size =1)
n2 = np.random.normal(loc=0, scale =1, size =1)
```

```python
df = pd.DataFrame({"n1":n1, "n2":n2, "sum":n1+n2})
```

```python
plt.figure(figsize=(12,7))
```

```python
plt.subplot(2,2,1)
plt.hist(df["n1"], label = "n1")
plt.title("1st number")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,2,2)
plt.hist(df["n2"], label = "n2", color = "green")
plt.title("2nd number")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,2,3)
plt.hist(df["sum"], label = "n1+n2", color = "red")
plt.title("Sum of two number")
plt.legend()
plt.tight_layout()
plt.show()
```
